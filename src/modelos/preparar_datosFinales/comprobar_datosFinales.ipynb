# Comprobación de datos para el dataset_final

```python
import pandas as pd
import os
import pyarrow.parquet as pq

# Definimos la ruta
parquet_path = os.path.join("..", "..", "..", "data", "processed", "tlc_clean", "datos_final.parquet")

if os.path.exists(parquet_path):
    # 1. Abrimos el archivo como un objeto de Parquet (esto no carga los datos)
    pfile = pq.ParquetFile(parquet_path)
    
    # 2. Obtenemos metadatos generales
    total_filas = pfile.metadata.num_rows
    print(f"Dimensiones totales del dataset: {total_filas:,} filas y {pfile.metadata.num_columns} columnas.")
    
    # 3. Leemos solo los bloques necesarios (el primero y el último)
    # read_row_group(0) lee el bloque inicial, el último es num_row_groups - 1
    df_inicio = pfile.read_row_group(0).to_pandas()
    df_fin = pfile.read_row_group(pfile.num_row_groups - 1).to_pandas()
    
    print(f"\n--- PRIMERAS FILAS (Inicio del periodo) ---")
    display(df_inicio.head(5))
    
    print(f"\n" + "="*50 + "\n")
    
    print(f"--- ÚLTIMAS FILAS (Final del periodo) ---")
    display(df_fin.tail(5))
    
    # Análisis ligero de una columna específica (ej: tipo_vehiculo)
    # Usamos value_counts() sobre un bloque para ver qué hay, sin cargar todo
    print(f"\n--- Muestra de 'tipo_vehiculo' (basada en el primer bloque) ---")
    print(df_inicio['tipo_vehiculo'].value_counts())

else:
    print(f"Error: No se encontró el archivo en la ruta relativa: {parquet_path}")
```

```python
print(f"--- Primeras 20 filas de 'tipo_vehiculo' (Bloque inicial) ---")
print(df_inicio['tipo_vehiculo'].head(20))
```

# Comprobación de datos para datos_final 

Estos son los datos limpios, con los que trabajaremos en los modelos 

```python
import pandas as pd
import os
import pyarrow.parquet as pq

# Ruta al archivo limpio que generamos
parquet_path = os.path.join("..", "..", "..", "data", "processed", "tlc_clean", "datos_final.parquet")

if os.path.exists(parquet_path):
    # 1. Inspección de metadatos (sin cargar el archivo a RAM)
    pfile = pq.ParquetFile(parquet_path)
    print(f"Archivo encontrado: {os.path.basename(parquet_path)}")
    print(f"Total de registros: {pfile.metadata.num_rows:,}")
    
    # 2. Cargar una muestra representativa para verificar la limpieza
    df_muestra = pfile.read_row_group(0).to_pandas()
    
    print("\n--- COMPROBACIÓN DE LA LIMPIEZA ---")
    
    # A) ¿tipo_pago ha desaparecido?
    if 'tipo_pago' not in df_muestra.columns:
        print("Columna 'tipo_pago': ELIMINADA correctamente.")
    else:
        print("Error: 'tipo_pago' sigue presente.")
        
    # B) ¿Quedan nulos en las columnas críticas?
    nulos_prop = df_muestra['propina'].isna().sum()
    nulos_pas = df_muestra['num_pasajeros'].isna().sum()
    
    print(f"Nulos en 'propina': {nulos_prop}")
    print(f"Nulos en 'num_pasajeros': {nulos_pas}")
    
    # 3. Vista rápida de datos
    print("\n--- VISTA PREVIA (Primeras 5 filas) ---")
    display(df_muestra[['propina', 'num_pasajeros']].head())
    
    # 4. Estadísticas rápidas para ver si la imputación funcionó
    print("\n--- ESTADÍSTICAS MUESTRALES ---")
    print(df_muestra[['propina', 'num_pasajeros']].describe())

else:
    print(f"Error: No se encontró el archivo en: {parquet_path}")
```

Comprobar si el dataset está ordenado cronológicamente

```python
# --- COMPROBACIÓN DE ORDEN CRONOLÓGICO ---
print("\n--- COMPROBACIÓN DE ORDEN ---")

# Asumiendo que tu columna de fecha se llama 'fecha_inicio'
# Si tiene otro nombre (ej: 'pickup_datetime'), cámbialo abajo:
columna_fecha = 'fecha_inicio' 

if columna_fecha in df_muestra.columns:
    # Verificamos si los datos del primer bloque están ordenados
    esta_ordenado = df_muestra[columna_fecha].is_monotonic_increasing
    
    if esta_ordenado:
        print(f"El dataset está ordenado cronológicamente por '{columna_fecha}'.")
    else:
        print(f"El dataset NO parece estar ordenado por '{columna_fecha}'.")
        print(f"Primeras fechas del bloque:\n{df_muestra[columna_fecha].head()}")
else:
    print(f"No se encontró la columna '{columna_fecha}' para verificar el orden.")
```

```python
# --- VISTA EXTENDIDA: PRIMERAS 30 FILAS ---
print("\n--- VISTA EXTENDIDA (Primeras 30 filas) ---")

# En lugar de cargar todo el archivo, leemos solo el primer grupo de filas (row group)
# y extraemos las primeras 30
df_muestra_30 = pfile.read_row_group(0).to_pandas().head(30)

# Mostramos el resultado
display(df_muestra_30)
```

necesitamos ver qué valores tienen algunas columnas específicas

```python
#Inspección de variables temporales y cíclicas
cols_temporales = ['hora', 'dia_semana', 'es_fin_semana', 'franja_horaria', 'hora_sen', 'hora_cos']

print("--- VALORES ÚNICOS Y EJEMPLOS DE VARIABLES TEMPORALES ---")

for col in cols_temporales:
    if col in df_muestra.columns:
        # Obtenemos los valores únicos (ordenados si son pocos)
        unique_vals = sorted(df_muestra[col].unique())
        
        print(f"\nVARIABLE: {col}")
        print(f"  > Tipo de dato: {df_muestra[col].dtype}")
        
        if len(unique_vals) <= 12: # Si son pocos (días, franjas, binarios)
            print(f"  > Valores únicos: {unique_vals}")
        else:
            print(f"  > Rango de valores: {min(unique_vals)} a {max(unique_vals)}")
            print(f"  > Ejemplo de las primeras 3 filas:\n{df_muestra[col].head(3).to_list()}")
    else:
        print(f"\nERROR: La columna {col} no se encuentra en el DataFrame.")

# Visualización en tabla para ver la relación entre hora y sus sen/cos
print("\n--- RELACIÓN HORA vs TRANSFORMACIÓN CÍCLICA ---")
display(df_muestra[['hora', 'hora_sen', 'hora_cos']].drop_duplicates().sort_values('hora').head(10))
```
