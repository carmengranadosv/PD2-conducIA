```python
import pandas as pd
import os

# Definimos la ruta de forma relativa al archivo Parquet
# Subimos 3 niveles: preparar_datosFinales -> modelos -> src -> raíz (PD2-conducIA)
parquet_path = os.path.join("..", "..", "..", "data", "processed", "tlc_clean", "dataset_final.parquet")

# Verificamos si el archivo existe antes de intentar leerlo
if os.path.exists(parquet_path):
    # Cargamos el dataset usando read_parquet
    df_verificacion = pd.read_parquet(parquet_path)
    
    # Nota: No hace falta pd.to_datetime porque Parquet mantiene el tipo datetime
    
    print(f"--- PRIMERAS FILAS (Inicio del periodo) ---")
    display(df_verificacion.head(5))
    
    print(f"\n" + "="*50 + "\n")
    
    print(f"--- ÚLTIMAS FILAS (Final del periodo) ---")
    display(df_verificacion.tail(5))
    
    print(f"\nDimensiones totales del dataset: {df_verificacion.shape[0]:,} filas y {df_verificacion.shape[1]} columnas.")
else:
    print(f"Error: No se encontró el archivo en la ruta relativa: {parquet_path}")
    print(f"Ruta absoluta intentada: {os.path.abspath(parquet_path)}")
```
